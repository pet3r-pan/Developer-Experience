```python
import pandas as pd    
import numpy as np
import glob


file_ids = []
file_names = []
dfs = []


lines = []  # store for your lines
files = glob.glob("../Interviews/*.txt")
for fle in files:
    print(fle.split("/")[-1]);
    file_name = fle.split("/")[-1]
    file_id = file_name.split(" ")[0]
    file_ids.append(file_id)
    file_names.append(" ".join(file_name.split(" ")[1:]))    
    
    txt = open("../Interviews/"+file_name, "r")
    indentation_level = [len(raw2) - len(raw2.lstrip(' ')) for raw2 in txt]
    txt = open("../Interviews/"+file_name, "r")
    txt_data = [raw.strip('\r\n\xef\xbb\xbf').replace(';', '') for raw in txt]

    df = pd.DataFrame(
            {'indentation_level': indentation_level,
             'txt_data': txt_data
            })
    df['level_0'] = np.where(df['indentation_level']==0, df['txt_data'], None)
    df = df.fillna(method='ffill')
    df = df[['level_0', 'txt_data', 'indentation_level']]
    df2 = df.groupby(['level_0'])['txt_data'].apply(' \r\n'.join).reset_index()
    df2['file_name'] = file_name
    df2['file_id'] = file_id
    df2 = df2[['file_id', 'file_name', 'level_0', 'txt_data']]
    dfs.append(df2) 
```

```python
df_full = pd.concat(dfs, axis=0)
df_full.to_csv("interview_data.csv", header=True, index=False)
```
